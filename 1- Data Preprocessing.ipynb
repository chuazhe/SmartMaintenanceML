# Feature Extraction


```python
# Create a dataframe to hold the last cycle for each turbine engine in the dataset
df_max_cycle = pd.DataFrame(train_data.groupby('id')['cycle'].max())
df_max_cycle.reset_index(level=0, inplace=True)
df_max_cycle.columns = ['id', 'last_cycle']
    
# Add Remaining Useful Life (RUL) as a new column for regression label
train_data = pd.merge(train_data, df_max_cycle, on='id')
train_data['RUL'] = train_data['last_cycle'] - train_data['cycle']
train_data.drop(['last_cycle'], axis=1, inplace=True)
```

```python
train_data.dtypes
```

```python
# Save to CSV
train_data.to_csv('Dataset/train.csv', index=False)
print("Train Data saved as Dataset/train.csv")
```

```python
# For Test data

# Create a dataframe to hold the last cycle for each turbine engine in the dataset
df_tst_last_cycle = pd.DataFrame(test_data.groupby('id')['cycle'].max())
df_tst_last_cycle.reset_index(level=0, inplace=True)
df_tst_last_cycle.columns = ['id', 'last_cycle']

# Add Remaining Useful Life (RUL) as a new column for regression label   
test_data = pd.merge(test_data, df_tst_last_cycle, on='id')
test_data = test_data[test_data['cycle'] == test_data['last_cycle']]
test_data.drop(['last_cycle'], axis=1, inplace=True)
    
test_data.reset_index(drop=True, inplace=True)
    
test_data = pd.concat([test_data, truth_data], axis=1)

# return df_test_in
```
